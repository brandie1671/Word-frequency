```python
import pandas as pd
from nltk.corpus import stopwords
from collections import Counter

# Load your dataset into a DataFrame (replace 'your_dataset.csv' with the actual path)
data = pd.read_csv('covid_2021_1.csv')

# Convert all comments to lowercase
data['comment_text'] = data['comment_text'].str.lower()

# Remove punctuation and special characters
data['comment_text'] = data['comment_text'].str.replace('[^\w\s]', '')

# Remove stop words
stop_words = set(stopwords.words('english'))
data['comment_text'] = data['comment_text'].apply(lambda text: ' '.join(word for word in text.split() if word not in stop_words))
```

```python
from collections import Counter

# Assuming 'comment_text' is a column in a pandas DataFrame
word_counts = Counter()

for comment in data['comment_text']:
    words = comment.split()  # Split the comment into individual words
    for word in words:
        word_counts[word] += 1
```

```python
top_10_words = word_counts.most_common(10)
```

```python
words = [word for word, count in top_10_words]
counts = [count for word, count in top_10_words]

plt.bar(words, counts)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequent Words in Social Media Comments')
plt.xticks(rotation=45)
plt.show()
```
